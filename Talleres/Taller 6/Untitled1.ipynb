```python
#Importo los datos
datos = np.genfromtxt("running_dirty_nov.csv",names = True, delimiter=",")
```

```python
#Realizo los respectivos cambios de coordenadas y teniedo en cuenta que el radio de la tierra es 6371Km
x = 6371*sin(np.deg2rad(90 - datos['lat']))*cos(np.deg2rad(datos['lon']))
y = 6371*sin(np.deg2rad(90 - datos['lat']))*sin(np.deg2rad(datos['lon']))
z = 6371*cos(np.deg2rad(datos['lat']))
#Se deja el tiempo en unidades de Horas y los dejo en un arreglo
tm2H = []
ts2H = []
t = []
for i in range(len(datos["min"])):
    tm2H.append(datos["min"][i]/60)
for i in range(len(datos["seg"])):
    ts2H.append(datos["seg"][i]/360)
for i in range(len(datos["time"])):
    t.append(datos["time"][i]+tm2H[i]+ts2H[i])
```

```python
#Para tener un nuevo sistema de coordenadas se resta la posicion inicial a todo en X y Y
novoT = []
novoX = x - x[0]
novoY = y - y[0]
#La altura se puede despreciar
novoZ = z - z[0]
for i in range(len(t)):
    novoT.append(t[i] - t[0])
```
