```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import os
import smopy
```

```python
#Importo los datos
datos = np.genfromtxt("running_dirty_nov.csv",names = True, delimiter=",")
```

```python
#Realizo los respectivos cambios de coordenadas y teniedo en cuenta que el radio de la tierra es 6371Km
x = 6371*sin(np.deg2rad(90 - datos['lat']))*cos(np.deg2rad(datos['lon']))
y = 6371*sin(np.deg2rad(90 - datos['lat']))*sin(np.deg2rad(datos['lon']))
z = 6371*cos(np.deg2rad(datos['lat']))
#Se deja el tiempo en unidades de Horas y los dejo en un arreglo
tm2H = []
ts2H = []
t = []
for i in range(len(datos["min"])):
    tm2H.append(datos["min"][i]/60)
for i in range(len(datos["seg"])):
    ts2H.append(datos["seg"][i]/360)
for i in range(len(datos["time"])):
    t.append(datos["time"][i]+tm2H[i]+ts2H[i])
```

```python
#Para tener un nuevo sistema de coordenadas se resta la posicion inicial a todo en X y Y
novoX = x - x[0]
novoY = y - y[0]
#La altura se puede despreciar
novoZ = z - z[0]
novoT = t - t[0]
```

```python
#Vector unitario norte y oriente, Z se puede depsreciar
#Primero se hace el cambio de unidades y la referencia
latitud = np.radians(90 - 4.5981)
longitud = np.radians(-74.0758)
#Vectores Unitarios
UNorte = [cos(latitud)*cos(longitud),cos(latitud)*sin(longitud), -sin(latitud)]
UOriente = [-sin(longitud), cos(longitud), 0]
#Se calculan los valores locales de X,Y y Z
VNorte = zeros(len(novoX))
for i in range(len(novoX)):
    LocalXN = UNorte[0]*novoX[i]
    LocalYN = UNorte[1]*novoY[i]
    LocalZN = UNorte[2]*novoZ[i]
    VNorte[i] = LocalXN + LocalYN + LocalZN
VOriente = zeros(len(novoY))
for i in range(len(novoY)):
    LocalXO = UOriente[0]*novoX[i]
    LocalYO = UOriente[1]*novoY[i]
    LocalZO = UOriente[2]*novoZ[i]
    VOriente[i] = LocalXO + LocalYO + LocalZO
    
#Para generar el plano en smopy se declara la referencia
refx = []
refy = []
for i in range(len(VNorte)):
    refy.append(VNorte[i] + 4.5981)
for i in range(len(VOriente)):
    refx.append(VOriente[i] + (-74.0758))
#Se llama a Smopy
map = smopy.Map(4.5981,-74.0758, z=11)
a,b = map.to_pixels(datos['lat'],datos['lon'])
plt.figure(figsize=(15,14))
map.show_mpl()
plt.plot(a, b, '-k', lw=1.4)
plt.plot(a[-1], b[-1], 'or', ms=1)
```

```python
plt.figure(1, figsize = (6, 10))
plt.plot(VOriente, VNorte)
```

```python
#Definir dx, dy y la gráfica
#Se inicia en cero el primer numero, que que al restar en el arreglo, el primer numero seria el ultimo-0, entonces la grafica quedaria mal
dx = VOriente - np.roll(VOriente, 1)
dy = VNorte - np.roll(VNorte, 1)
dx[0] = 0
dy[0] = 0
distancia = []
#Defino la distancia
for i in range(len(VNorte)):
    distancia.append(np.sqrt(dx[i]**2 + dy[i]**2))
#Ahora se sum cada miempre de distancia con el anterior
sumaTotal = np.cumsum(distancia)
plt.plot(novoT,sumaTotal)
```
